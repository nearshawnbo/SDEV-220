```python
#4.1
import random
secret = random.randint(1,10)
guess = random.randint(1,10)
if guess < secret:
    print("Too Low")
elif guess > secret:
    print("too high")
else:
    print("just right")
```

```python
#4.2
small = True
green = True

# cherry
if small and not green:
    print("cherry")
else:
    print("cherry does not match")

# pea
if small and green:
    print("pea")
else:
    print("pea does not match")

# watermelon
if not small and green:
    print("watermelon")
else:
    print("watermelon does not match")

# pumpkin
if not small and not green:
    print("pumpkin")
else:
    print("pumpkin does not match")
```

```python
#6.2
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:  # number > guess_me
        print("oops")
        break
    number += 1
```

```python
#6.3
guess_me = 5

for number in range(10):
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:  # number > guess_me
        print("oops")
        break
```
